# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:
1. whale_returns.csv
2. algo_returns.csv
3. sp500_history.csv

## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Reading whale returns
whale_returns_csv = Path("Resources/whale_returns.csv")
whale_returns_df = pd.read_csv(whale_returns_csv, index_col='Date', parse_dates=True, 
                               infer_datetime_format=True)
whale_returns_df.sort_index(ascending=True, inplace=True)
```

```python
# Count nulls
whale_returns_df.isnull().sum()
```

```python
# Drop nulls
whale_returns_df.dropna(inplace=True)
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
algo_returns_csv = Path("Resources/algo_returns.csv")
algo_returns_df = pd.read_csv(algo_returns_csv, index_col='Date', parse_dates=True, 
                              infer_datetime_format=True)
algo_returns_df.sort_index(ascending=True, inplace=True)
```

```python
# Count nulls
algo_returns_df.isnull().sum()
```

```python
# Drop nulls
algo_returns_df.dropna(inplace=True)
```

## S&P 500 Returns

Read the S&P500 Historic Closing Prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices, sorting index
sp500_history_csv = Path("Resources/sp500_history.csv")
sp500_history_df = pd.read_csv(sp500_history_csv, index_col='Date', parse_dates=True, 
                                 infer_datetime_format=True)
sp500_history_df.sort_index(ascending=True, inplace=True)
```

```python
# Check Data Types
sp500_history_df.dtypes
```

```python
# Fix Data Types
sp500_history_df['Close'] = sp500_history_df['Close'].str.replace('$','').astype(float)
sp500_history_df.dtypes
```

```python
# Calculate Daily Returns
sp500_returns_df = sp500_history_df.pct_change()
```

```python
# Drop nulls
sp500_returns_df.dropna(inplace=True)
```

```python
# Rename Column
sp500_returns_df.columns = ['S&P 500']
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Concatenate all DataFrames into a single DataFrame
combined_returns_df = pd.concat([whale_returns_df, sp500_returns_df, algo_returns_df], axis='columns', join='inner')
```

Calculate and Plot the daily returns and cumulative returns.

## Risk

Determine the _risk_ of each portfolio:

1. Create a box plot for each portfolio. 
2. Calculate the standard deviation for all portfolios
4. Determine which portfolios are riskier than the S&P 500
5. Calculate the Annualized Standard Deviation

```python
# Box plot to visually show risk
combined_returns_df.plot(kind = "box", figsize = (20,10))
```

```python
# Daily Standard Deviations
# Calculate the standard deviation for each portfolio. Which portfolios are riskier than the S&P 500?
combined_std_df = combined_returns_df.std().to_frame()
combined_std_df.columns = ['STD']
```

```python
# Determine which portfolios are riskier than the S&P 500
higher_than_sp500_std = combined_std_df[combined_std_df['STD'] > combined_std_df.loc['S&P 500', 'STD']]
print(highter_than_sp500_std.columns = ['Highter STD than for S&P 500'])
```

```python
# Calculate the annualized standard deviation (252 trading days)
annual_combined_std_df = combined_std_df * np.sqrt(252)
annual_combined_std_df.columns = ['Annual STD']
annual_combined_std_df.sort_index(ascending=False, inplace=True)
annual_combined_std_df
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Calculate and plot the rolling standard deviation for the S&PP 500 using a 21 day window
2. Calculate the correlation between each stock to determine which portfolios may mimick the S&P 500
2. Calculate and plot a 60 day Beta for Berkshire Hathaway Inc compared to the S&&P 500

```python
# Calculate and plot the rolling standard deviation for the S&PP 500 using a 21 day window
sp500_rolling_std_21 = combined_returns_df[['S&P 500']].rolling(window=21).std() 
sp500_rolling_std_21.plot(figsize=(20,10), title='21-Day Rolling STD of S&P 500')
```

```python
# Correlation
correlation_sp500 = pd.DataFrame(combined_returns_df.corr().loc[:,'S&P 500'])
correlation_sp500.sort_values(by='S&P 500', ascending=False, inplace=True)
correlation_sp500
```

```python
# Calculate Beta for a single portfolio compared to the total market (S&P 500)
bhi_rolling_covariance = combined_returns_df['BERKSHIRE HATHAWAY INC'].rolling(window=60).cov(combined_returns_df['S&P 500'])
rolling_variance = combined_returns_df['S&P 500'].rolling(window=60).var()
rolling_beta = bhi_rolling_covariance / rolling_variance
rolling_beta.plot(figsize=(20, 10), title='60-Day Rolling Beta of BHI')
```

## Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. (After all, if you could invest in one of two portfolios, each offered the same 10% return, yet one offered lower risk, you'd take that one, right?)

Calculate and plot the annualized Sharpe ratios for all portfolios to determine which portfolio has the best performance

```python
# Annualzied Sharpe Ratios
sharpe_ratios = (combined_returns_df.mean()*252/(combined_returns_df.std()*np.sqrt(252))).to_frame()
sharpe_ratios.columns = ['Sharpe Ratio']
sharpe_ratios
```

 plot() these sharpe ratios using a barplot.
 On the basis of this performance metric, do our algo strategies outperform both 'the market' and the whales?

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot(kind = "bar", figsize=(5,5), title = "Sharpe Ratios")
```

## Choose 3-5 custom stocks with at last 1 year's worth of historic prices and create a DataFrame of the closing prices and dates for each stock.

```python
# Read the first stock
aapl_data = pd.read_csv(Path('Resources/aapl_data.csv'), index_col='Date', 
                        infer_datetime_format=True, parse_dates = True, decimal=',')
aapl_data.sort_index(inplace=True)
```

```python
# Read the second stock
goog_data = pd.read_csv(Path('Resources/goog_data.csv'), index_col='Date', 
                        infer_datetime_format=True, parse_dates = True, decimal=',')
goog_data.sort_index(inplace=True)
```

```python
# Read the third stock
fb_data = pd.read_csv(Path('Resources/fb_data.csv'), index_col='Date', 
                        infer_datetime_format=True, parse_dates = True, decimal=',')
fb_data.sort_index(inplace=True)
```

```python
# Concatenate all stocks into a single DataFrame
stocks = pd.concat([aapl_data, goog_data, fb_data], axis = 'columns', join='inner')
stocks.columns = ['AAPL', 'GOOG', 'FB']
stocks.sort_index(inplace=True)
```

```python
# Drop Nulls
stocks.dropna(inplace=True)
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Calculate weighted portfolio returns
weights = [1/3, 1/3, 1/3]
my_portfolio = stocks.dot(weights)
my_portfolio_daily = my_portfolio.pct_change()
my_portfolio_daily.dropna(inplace=True)
my_portfolio_daily.head()
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
all_portfolios = pd.concat([my_portfolio_daily,combined_returns_df], axis='columns', join ='inner')
all_portfolios.rename(columns = {0:"My Portfolio"}, inplace=True)
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)
all_portfolios.dropna(inplace=True)
```

## Re-run the performance and risk analysis with your portfolio to see how it compares to the others

```python
# Risk
all_portfolios_std = pd.DataFrame(all_portfolios.std()).rename(columns = {0:'STD'}) * np.sqrt(252)
all_portfolios_std.sort_values(by='STD', inplace=True, ascending=False)
all_portfolios_std
```

```python
# Rolling
all_portfolios.rolling(window=21).std().plot(figsize=(25,10), title='21-Day Rolling STD')
```

```python
# Beta
covariance_my_portfolio = all_portfolios['My Portfolio'].cov(all_portfolios['My Portfolio'])
variance = combined_returns_df['S&P 500'].var()
my_portfolio_beta = covariance_my_portfolio / variance
my_portfolio_beta 
```

```python
rolling_covariance = all_portfolios['My Portfolio'].rolling(window=30).cov(all_portfolios['S&P 500'])
rolling_variance = all_portfolios['S&P 500'].rolling(window=30).var()
rolling_beta = rolling_covariance / rolling_variance
rolling_beta.plot(figsize=(20,10), title='30-Day Rolling Beta')
```

```python
# Annualzied Sharpe Ratios
all_sharpe_ratios = (all_portfolios.mean()* 252)/ (all_portfolios.std() * np.sqrt(252))
all_sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
all_sharpe_ratios.plot(kind="bar",figsize=(5,5), title="Sharpe Ratios")
```
